```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import csv
import pandas as pd
import numpy as np
import cv2 as cv
%matplotlib inline
```

```python
data = pd.read_csv('./results/vocab.csv', delimiter='\t')
v = data['Vocabsize'].to_numpy() / 1000
acc = data['Accuracy'].to_numpy()

plt.style.use('seaborn')
plt.plot(v, acc, color='purple', marker='o')
plt.title('vocabulary size (k) and accuracy')

for x, y in zip(v, acc):
    plt.text(x+1, y, '%d'%x)

# plt.savefig('./v.jpg', dpi=600)
plt.show()
```

```python
data = pd.read_csv('./results/category.csv', delimiter='\t')
xs = data['Category']
ys = data['Accuracy']
plt.barh(xs, ys, color='purple')
for x, y in zip(xs, ys):
    plt.text(y, x, y)
plt.title('Bag of SIFT Representation')
# plt.savefig('./c.jpg', dpi=600)
plt.show()
```

```python
import os
import random
train_dirs = os.listdir('./train')
cats = [s.lower() for s in train_dirs]
train_filenames = {}
for dirname, cat in zip(train_dirs, cats):
    dirpath = os.path.join('./train', dirname)
    train_filenames[cat] = [os.path.join(dirpath, name) for name in os.listdir(dirpath)]
```

```python
forest = train_filenames['forest']
suburb = train_filenames['suburb']

random.shuffle(forest)
random.shuffle(suburb)
forest = forest[:4]
suburb = suburb[:4]

plt.figure(figsize=(5, 5))
sift = cv.SIFT_create()
for i in range(4):
    img = cv.imread(forest[i])
    kp = sift.detect(img,None)
    kp, des = sift.compute(img, kp)
    img2 = cv.drawKeypoints(img, kp, None, color=(100,30,100))
    plt.subplot(2, 2, i+1)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(img2)
# plt.savefig('./forest.jpg', dpi=600)
```

```python
plt.figure(figsize=(10, 6))
sift = cv.SIFT_create()
for i in range(4):
    img = cv.imread(suburb[i])
    
    kp = sift.detect(img,None)
    kp, des = sift.compute(img, kp)
    img2 = cv.drawKeypoints(img, kp, None, color=(100,30,100))
    
    plt.subplot(2, 2, i+1)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(img2)
# plt.savefig('./suburb.jpg', dpi=600)
```
